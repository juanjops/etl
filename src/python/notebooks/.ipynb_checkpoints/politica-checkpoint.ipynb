```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
demografia = pd.read_csv("03002bsc.csv", encoding="latin-1", delimiter=";")
demografia["Provincias"] = demografia["Provincias"].apply(lambda x: float(x[0:2]))
demografia.rename(columns = {"Provincias": "Provincia",
                            "Edad (grupos quinquenales)": "Edad",
                            "Espanoles/Extranjeros": "Inmigracion"}, inplace = True)
demografia.head()
```

```python
demografia.Inmigracion.value_counts()
```

```python
state_office = demografia.groupby(['Provincia', 'Inmigracion', "Sexo", "Edad"]).agg({'Total': 'sum'})
c = state_office / state_office.groupby(level=[0, 1, 2]).transform("sum")
c.reset_index(inplace=True)
c.rename(columns = {"Total": "Porcentaje_Poblacion"}, inplace = True)
c["Porcentaje_Poblacion"] = c["Porcentaje_Poblacion"].apply(lambda x: x*100)
```

```python
plt.figure(figsize=(20,10))

def plot_function(data_set, c, partido, edad, inmigracion, sexo):
    
    data_filtered = data_set[(data_set.Partido==partido)]
                             
    data_with_means = pd.merge(data_filtered, c, on=['Provincia', 'Inmigracion', "Sexo", "Edad"])
                             
    data_with_means.replace({"Provincia": provincias}, inplace=True)
    
    ax = data_with_means[(data_with_means.Edad==edad)
                    & (data_with_means.Inmigracion==inmigracion)
                    & (data_with_means.Sexo==sexo)].set_index('Porcentaje_Poblacion')['Porciento'].plot(style='o')

    def label_point(x, y, val, ax):
        a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
        for i, point in a.iterrows():
            ax.text(point['x'], point['y'], str(point['val']))

    label_point(data_with_means[(data_with_means.Edad==edad)
                    & (data_with_means.Inmigracion==inmigracion)
                    & (data_with_means.Sexo==sexo)].Porcentaje_Poblacion, 
                data_with_means[(data_with_means.Edad==edad)
                    & (data_with_means.Inmigracion==inmigracion)
                    & (data_with_means.Sexo==sexo)].Porciento,
                data_with_means[(data_with_means.Edad==edad)
                    & (data_with_means.Inmigracion==inmigracion)
                    & (data_with_means.Sexo==sexo)].Provincia, ax)

    
plot_function(data_set, c, "Partido Popular (PP)", "30-34 anos", "Extranjeros", "Ambos sexos")

plot_function(data_set, c, "Vox (VOX)", "30-34 anos", "Extranjeros", "Ambos sexos")
```
