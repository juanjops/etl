```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data_frame = (pd.read_csv("C:\\Users\\Sectorea\\Code\\database_linkedin\\text_analyze_data_scientist_London_Past_24_hours_2020-04-12.csv"))
data_frame.shape
```

```python
data_frame.head()
```

```python
def get (feature): 
    variable_extension = " ".join(review for review in feature if type(review) is not float)
    wordcloud = WordCloud(
        max_font_size=50, max_words=100, background_color="white",collocations=False,
        prefer_horizontal=1).generate(variable_extension)
    variable_count = variable_extension.split(" ")
    print(Counter(variable_count).most_common(100))
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
get(data_frame.key_words)
```
